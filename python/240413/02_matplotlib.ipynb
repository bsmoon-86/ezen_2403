```python
# 라이브러리 로드 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np 
import platform
```

```python
# matplot을 이용하여 라인 그래프를 생성 
# y축의 데이터만 존재해도 그래프 생성이 가능
y_data = [10, 5, 8, 16]
plt.plot(y_data)
# show() : print()같은 기능
plt.show()
```

```python
# 라인 그래프에 x축 데이터, y축 데이터를 모두 대입
x_data = [1, 3, 6, 10]
y_data = [1, 2, 3, 4]
plt.plot(x_data, y_data)
plt.show()
```

```python
# 라인 그래프 라인 색상을 변경 
plt.plot(x_data, y_data, '#FF00FF')
plt.show()
```

```python
# 라인의 종류를 변경 
plt.plot(x_data, y_data, 'r--')
plt.show()
```

```python
## 시각화중 한글을 이용하는 경우
# matplotlib의 폰트를 변경

# 현재 컴퓨터의 os가 무엇인가?
if platform.system() == 'Darwin' :
    # 현재 컴퓨터의 os가 MAC인 경우 
    plt.rc('font', family = 'AppleGothic')
else:
    plt.rc('font', family='Malgun Gothic')
```

```python
## 그래프에 x축의 이름, y축의 이름, 그래프의 이름 대입 
plt.plot(x_data, y_data)
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title('그래프')
plt.show()
```

```python
# 그래프에 범례 추가 
plt.plot(x_data, y_data, label='A')
plt.plot(x_data, [8, 5, 3, 1], label='B')
plt.legend(loc='upper left')
plt.show()
```

```python
## 막대 그래프 생성 
## x축 데이터, y축 데이터가 필수 항목
x = [1,2,3,4]
y = [80, 70, 85, 90]

plt.bar(x, y)
plt.show()
```

```python
# 누운 막대 그래프 
plt.barh(x, y)
plt.show()
```

```python
## 옵션 추가 ( 막대의 색상을 변경, 막대의 너비를 변경 )
plt.bar(x, y, color=['r', 'k'], width=0.1)
plt.show()
```

```python
graph = plt.bar(x, y)
graph[2].set_color('r')
plt.show()
```

```python
df = pd.read_csv('../csv/drinks.csv')
```

```python
# pandas에서 상관계수를 출력
# 상관 계수 : -1부터 1까의 수치로 표현 
    # -1 : 하나의 데이터가 상승하면 다른 하나의 데이터가 감소
    # 1 : 하나의 데이터가 상승하면 다른 하나의 데이터도 상승
    # 0 : 두개의 데이터가 무작위
corr = df.loc[:, 'beer_servings' : 'total_litres_of_pure_alcohol'].corr('pearson')
```

```python
# seaborn 라이브러리를 이용하여 heatmap 생성

# x, y의 이름을 축약 해서표시하기 위한 데이터를 생성
cols = ['beer', 'spirit', 'wine', 'alcohol']

# heatmap 생성
hm = sns.heatmap(
    corr.values, 
    # 사이드바 표시 유무
    cbar = True, 
    # 히트맵에 숫자를 표시 유무
    annot = True, 
    # 히트맵을 정사각형을 표시 유무
    square= True, 
    # 숫자의 소수점 자리수 지정 
    fmt = '0.3f', 
    # 숫자의 크기 
    annot_kws={'size' : 20}, 
    xticklabels=cols, 
    yticklabels=cols
)
plt.show()
```

```python
sns.pairplot(
    df.loc[:, 'beer_servings' : 'total_litres_of_pure_alcohol'], 
    kind='reg', 
    corner=False
)
plt.show()
```
